```python
# Importing libraries and necessary packages. 
import pandas as pd 
import numpy as np 
import summarytools
from summarytools import dfSummary
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
#importing data set
md = pd.read_csv('/Users/brittles/Bioinformatics/medical_market_basket.csv')
md.head(5)
```

```python
#checking the data types of each variable and for any missing values
md.info()
```

```python
print(md.shape)
```

```python
dfSummary(md)
```

```python
md_cleaning = md[md['Presc01'].notna()]
```

```python
print(md_cleaning.shape)
```

```python
#create a list to prepare data for encoding
trans = []

for i in range (0,7501):
    trans.append([str(md_cleaning.values[i,j])for j in range (0,20)])
  
```

```python
#transcationalize dataset
TE = TransactionEncoder() 
array = TE.fit(trans).transform(trans)
```

```python
#putting back into dataframe
md_clean= pd.DataFrame(array, columns = TE.columns_)
md_clean
```

```python
for col in md_clean.columns:
    print(col)
```

```python
md_clean = md_clean.drop(['nan'], axis = 1)
```

```python
for col in md_clean.columns:
    print(col)
```

```python
#saving to csv
#md_clean.to_csv('md_clean_mba_D212.csv', index=False)
```

```python
#mining rules by applying apriori
a_rules = apriori(md_clean, min_support = 0.05, use_colnames = True )
a_rules.head()
```

```python
#set metrics to mine rules 
rules = association_rules(a_rules, metric = 'lift', min_threshold = 1)
rules
```

```python
# Identify the top 3 rules by sorting the association table based on lift
top_rules = rules.sort_values(by='lift', ascending=False).head(3)

# print the sorted table with variables of interest
print(top_rules[['antecedents', 'consequents','lift', 'confidence','support']])
```
